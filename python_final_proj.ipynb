# Outliers

```python
sns.boxplot(data=df ,y='Salary')
plt.title('Salary')
plt.show()
```

```python
 # Inter Quantile Range
```

```python
# calc IQR 
Q3=df['Salary'].quantile(0.75)
Q1=df['Salary'].quantile(0.25)
IQR=Q3-Q1
IQR
```

```python
# cal upper outlier
upper_outlier= Q3 + (1.5 * IQR)
upper_outlier
```

```python
# cal lower outlier
lower_outlier= Q1 - (1.5 * IQR)
lower_outlier
```

```python
# dealing with lower outlier
df = df[df['Salary'] >= 0]
```

```python
df.describe().T
```

```python
# dealing with upper outlier
sns.boxplot(data=df,x='Education',y='Salary')
plt.title('Education VS Salary')
plt.show()
# doctores are most salaries so we don't need to remove them
```
